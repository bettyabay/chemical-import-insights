# **NLP Brand Categorization**

```python
import re
from nltk.tokenize import word_tokenize
import nltk
```

```python
#ensure that the punkt package is downloaded.
nltk.download('punkt')
```

Columns of Interest for brand categorization

```python
#selecting specific columns
columns_of_interest = ['Trader', 'HS Code' , 'HS Description', 'Commercial / Brand Name', 'Reg. date' , '# of packages']
selected_data = data_import[columns_of_interest]
```

```python
# Define expanded category mapping with keywords
category_mapping = {
    'Construction': [
        'BOND', 'SEAL', 'PRIMER', 'PAINT', 'CONSTRUCTION', 
        'CEMENT', 'FIBER DRUMS', 'WALL PAPER', 'ADHESIVE', 
        'SEALANT', 'MODIFIER'
    ],
    'Chemicals': [
        'ACID', 'CHEMICAL', 'SOLVENT', 'REAGENT', 
        'POVIDONE', 'KOVIDONE', 'THICKENER', 'AGENT', 
        'DIVERGAN', 'FORMULATION', 'MIXTURE'
    ],
    'Industrial': [
        'INDUSTRIAL', 'MACHINERY', 'EQUIPMENT', 
        'FOAM ASSY', 'ENGINEERED BASE STABILISER', 
        'COATING', 'ADDITIVE'
    ],
    'Plastics and Composites': [
        'POLYVINYL', 'POLYPROPYLENE', 'PVC', 'COPOLYMER', 
        'HDPE', 'LDPE', 'POLYMER', 'RESIN', 'RAW MATERIAL'
    ],
    'Agricultural': [
        'AGRI', 'FERTILIZER', 'SEED', 'HARVEST', 
        'HERBICIDE', 'DISPERSIBLE', 'EMULSION', 
        'LIQUID', 'POWDER', 'GRANULE'
    ],
    'Other': [
        'PRODUCT', 'MATERIAL', 'COMPOUND', 'SOLUTION', 
        'GENERAL', 'MIXTURE', 'ADHESIVE', 'SEALANT', 
        'FILLER', 'FORMULATION'
    ]
}
```

```python
# Function to preprocess and tokenize brand names
def preprocess_and_tokenize(text):
    text = text.upper()  # Convert to uppercase for consistency
    text = re.sub(r'[^A-Z\s]', '', text)  # Remove special characters
    tokens = word_tokenize(text)  # Tokenize the text
    return tokens
```

```python
# Function to categorize brand names based on tokens
def categorize_brand_name(brand_name):
    tokens = preprocess_and_tokenize(brand_name)
    for category, keywords in category_mapping.items():
        if any(keyword in tokens for keyword in keywords):
            return category
    return 'Other'  # Default category if no keywords match
```

```python
# Apply categorization
selected_data['Predict Category'] = selected_data['Commercial / Brand Name'].apply(categorize_brand_name)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
```

Handling Outliers

```python
# Calculate IQR for Quantity Imported
Q1 = data_model['# of packages'].quantile(0.25)
Q3 = data_model['# of packages'].quantile(0.75)
IQR = Q3 - Q1

# Define bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Remove outliers
data_model = data_model[
    (data_model['# of packages'] >= lower_bound) & 
    (data_model['# of packages'] <= upper_bound)
]
```

Let's organise columns of intereset in a dataset for predictive model training.
